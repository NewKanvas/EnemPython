# Palavras para trocar

1. **Também**

"A palavra 'também' é uma excelente palavra de ligação. No entanto, se você a utilizar em quase todas as frases, ela se destacará. Felizmente, existem outras palavras que você pode usar em seu lugar, tais como:

    Além disso
    Adicionalmente
    Além do mais

    Também
    Além disso

Outras palavras de ligação também estão disponíveis, mas costumam ser mais específicas. 'Da mesma forma', por exemplo, pode ser usado ao vincular coisas semelhantes, enquanto 'no entanto' introduz uma contrastação."

2. **Disse**

"Se você estiver citando várias fontes, pode se ver usando 'disse' ou 'diz' frequentemente. Para evitar repetições, tente usar estas alternativas:

    Afirmou
    Declarou
    Argumentou

    Sugeriu
    Explicou

Lembre-se de que algumas dessas são específicas para determinadas situações. Por exemplo, só faria sentido usar 'argumentou' se a pessoa citada estiver defendendo algo!

Vale ressaltar que repetir 'disse' nem sempre é um problema. Na verdade, a maioria dos escritores de ficção sugere manter 'disse' ou 'diz', pois usar muitos sinônimos pode ser distraente."

3. **Portanto**

"Usamos 'portanto' para mostrar que algo é resultado de algo mais. Mas, como ensaios podem envolver muitos links e conclusões, isso pode ser excessivo. Você pode querer usar alguns destes termos em seu lugar:

    Como resultado
    Consequentemente
    De acordo com isso

    Assim
    Portanto
    Por essa razão"

4. **Incluindo**

"Se você estiver apresentando um exemplo de algo que foi mencionado anteriormente, pode usar a palavra 'incluindo' (por exemplo, 'Eu conheço muitas palavras, incluindo...'). Outros termos podem ser usados da mesma forma, como 'como' e 'tipo' (embora 'tipo' seja bastante informal).

Você também pode introduzir exemplos com frases como 'por exemplo' ou 'por instance'."

5. **Interessante**

"Esperançosamente, seu ensaio será sobre algo que você ache interessante. A palavra 'interessante' em si, no entanto, é um tanto entediante por ser muito utilizada. Como tal, você pode precisar de um ou dois sinônimos:

    Notável
    Notável
    Marcante

    Intrigante
    Significativo
    Fascinante

Lembre-se, no entanto, que simplesmente observar algo como 'interessante' não é suficiente. Na escrita acadêmica, você também precisa explicar por que algo é interessante para mostrar seu entendimento."

6. **Há/Existem/É**

"Finalmente, temos 'há', 'existem' e 'é'. Essas frases são frequentemente usadas para começar frases, algo conhecido como 'construção expletiva' (e não porque estão cheias de palavrões). Nesse tipo de frase, o sujeito vem depois do verbo. Por exemplo:

'Há muitos problemas que podemos enfrentar.'

O sujeito acima é 'nós', enquanto o verbo principal é 'somos'. Em frases padrão, no entanto, o verbo deve vir depois do sujeito. Fazendo isso, podemos tornar nossa escrita mais concisa:

'Podemos enfrentar muitos problemas.'

Ao eliminar 'há', portanto, fazemos uma afirmação mais impactante."

# Conectivos
## Use este conetivos para ter uma redação melhor

**Conectores temporais:**
- Anteriormente;
- Posteriormente;
- Simultaneamente;
- Nesse ínterim.

**Conectores de causa e efeito:**
- Portanto;
- Assim;
- Por conseguinte;
- Como resultado;
- Em vista disso.

**Conectores de comparação:**
- Da mesma forma;
- Similarmente;
- À semelhança de;
- Assim como.

**Conectores de contraste:**
- No entanto;
- Por outro lado;
- Apesar disso;
- Em contrapartida.

**Conectores de adição:**
- Além disso;
- Adicionalmente;
- Acrescentando a isso;
- Também.

**Conectores de condição:**
- Se;
- Caso;
- Sob a condição de que;
- Contanto que.

**Conectores de explicação:**
- Isto é;
- Ou seja;
- Em outras palavras;
- Em razão disso.
